```python
import pandas as pd
from numpy import *
from scipy.optimize import *
kpi=5


whl=222.4 #wheel radius with inflated tyre
stl=100
m1=53
m2=53
g=9.81
theta=30.53475
phi=19.7152 
arm=[]
for i in range (40):
    kpi=round(kpi+0.1,2)
    y=tan(kpi*0.01745329251)
    x=whl*y
    sr=stl-x #scrub radius
    srm=(sr/1000) #scrub radius in meter
    w1=m1*g
    w2=m2*g
    ag=sin(kpi*0.01745329251)
    t1=sin(theta*0.01745329251)
    t2=sin(phi*0.01745329251)
    tiw=w1*sr*ag*t1 #torque on inner wheels
    tow=w2*sr*ag*t2 #torque on outer wheels
    arm.append([kpi,whl,stl,m1,m2,g,theta,phi,y,x,sr,srm,w1,w2,ag,t1,t2,tiw,tow])
    print(arm)
    
    
```

```python
df=pd.DataFrame(data=arm,columns=["kpi","wheel radius","stub lenght","m1","m2","g","theta","phi","tan kpi","x","scrub radius","scrub radius in mm","w1","w2","sin kpi","sin theta","sin phi","tiw","tow"])
```
